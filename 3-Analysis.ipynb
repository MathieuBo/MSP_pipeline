## Load nuclear segmentation

```python
# Reload segmentation
dat = np.load('../full_hipp_raw_seg.npy', allow_pickle=True).item()
```

```python
viewer.add_labels(dat['masks'], name='cellpose', scale=[6,1,1])
```

## Nuclei-based clustering

```python
regionprops(image_layer=viewer.layers['dapi'],
            labels_layer=viewer.layers['cellpose'],
            napari_viewer=viewer,
           size=True,
           intensity=True)
```

```python
nuclei_table = get_table(labels_layer=viewer.layers['cellpose'], viewer=viewer)
```

```python
data = pd.DataFrame(nuclei_table.get_content())
```

```python
# Remove smaller object that are not nuclei, set cutoff at 2000
data = data[data['area'] > 2000]
```

```python
sns.kdeplot(data['area'], shade=True)
plt.show()
```

```python
# UMAP dim red on nuclei data

fit = umap.UMAP(n_components=15)
%time u = fit.fit_transform(data[data.columns[1:-1]]) # removing index and label cols
```

```python
plt.scatter(u[:,0], u[:,1])
```

```python
umap_df = pd.DataFrame(u, columns=['UMAP0', 'UMAP1'])
```

```python
dapi_umap = pd.merge(data.reset_index(drop=True), umap_df, left_index=True, right_index=True)
```

## MC1 analysis

```python
regionprops(image_layer=viewer.layers['mc1'],
            labels_layer=viewer.layers['cellpose'],
            napari_viewer=viewer,
           size=False,
           intensity=True)
```

```python
mc1_table = get_table(labels_layer=viewer.layers['cellpose'], viewer=viewer)
```

```python
data_mc1 = pd.DataFrame(mc1_table.get_content())
```

```python
full_data = pd.merge(left=dapi_umap, right=data_mc1, on='index', how='inner', suffixes=('_dapi', '_mc1'))
```

```python
plt.figure(figsize=(5,4), constrained_layout=True)
sns.scatterplot(x='UMAP0', y='UMAP1', data=full_data, hue=full_data['mean_intensity_mc1']>450)

plt.savefig('E:mathieu/220222_full_hipp_IF/mc1.png', dpi=300)
plt.show()
```

## LF analysis

```python
regionprops(image_layer=viewer.layers['lf'],
            labels_layer=viewer.layers['cellpose'],
            napari_viewer=viewer,
           size=False,
           intensity=True)
```

```python
lf_table = get_table(labels_layer=viewer.layers['cellpose'], viewer=viewer)
```

```python
data_lf = pd.DataFrame(lf_table.get_content())
```

```python
full_data = pd.merge(left=full_data, right=data_lf, on='index', how='inner', suffixes=('', '_lf'))
```

```python
plt.figure(figsize=(6,4), constrained_layout=True)
plt.scatter(full_data['UMAP0'], full_data['UMAP1'], c=full_data['mean_intensity'],
            cmap='magma')
cbar = plt.colorbar()
cbar.set_label('Lipofuscin intensity', rotation=270)

plt.ylabel('UMAP1')
plt.xlabel('UMAP0')


plt.savefig('E:mathieu/220222_full_hipp_IF/lf.png', dpi=300)
plt.show()
```
